# **Chainlit Deploy**

```python
import subprocess

# Run the command as a background process
process = subprocess.Popen(['chainlit', 'run', 'app.py', '--port', '8000'],
                           stdout=open('./logs.txt', 'w'),
                           stderr=subprocess.STDOUT)
print(f"Chainlit is running with PID: {process.pid}")
```

```python
import time

def check_logs():
    success_message = "Your app is available at http://localhost:"
    log_file = "./logs.txt"

    while True:
        try:
            with open(log_file, 'r') as f:
                file_content = f.read()
                if success_message in file_content:
                    return "Success: Found the message in logs!"
        except FileNotFoundError:
            return f"Error: File {log_file} not found."
        except IOError as e:
            return f"Error: Failed to read file {log_file}: {e}"

        # Pause for a moment before checking again
        time.sleep(1)

# Continuously check logs until success message is found
while True:
    result = check_logs()
    if result.startswith("Success"):
        print(result)
        break
    elif result.startswith("Error"):
        print(result)
        break
```
